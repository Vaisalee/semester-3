imagine we have five nodes...cancer..smoking...absenstos consumption ...x-ray..blood vomiting..now we have a patient who by nature very less and was grown in an environment which has little bit of asbsetos around in the form of smoke or roof tops..he assumes he can develop cancer.during routine check ups there was something unusual in the x-ray shouting a tumor.unfortunately the patient had blood vomit.

```python
from protopunica import*
```

protopunica is a python library that implements probalistic models

```python
smoking=Node(DiscreteDistribution({"High smoking":0.7,"Low smoking":0.3}),name="smoking")
```

smoking is the root node and is not having any parent and has 2 values high consumption and low consumption

```python
asbes_consum=Node(DiscreteDistribution({"High cons":0.7,"Low cons":0.3}),name="asbes_consum")
```

asbestos consumption is also a root node which is not having any parent and has 2 values high consumption and low consumption

```python
cancer=Node(ConditionalProbabilityTable([
    ["High smoking","High cons","Pos",0.4],
    ["High smoking","High cons","Neg",0.6],
    ["High smoking","Low cons","Pos",0.3],
    ["High smoking","Low cons","Neg",0.7],
    ["Low smoking","Low cons","Pos",0.1],
    ["Low smoking","Low cons","Neg",0.9],
    ["Low smoking","High cons","Pos",0.02],
    ["Low smoking","High cons","Neg",0.98]
],[smoking.distribution,asbes_consum.distribution]),
            name="cancer")
```

```python
scan=Node(ConditionalProbabilityTable([
    ["Pos","scan_Pos",0.8],
    ["Pos","scan_Neg",0.2],
    ["Neg","scan_Pos",0.1],
    ["Neg","scan_Neg",0.9]
    
],[cancer.distribution]),name="scan")
```

```python
blood_vomiting=Node(ConditionalProbabilityTable([
    ["Pos","b.v_Pos",0.7],
    ["Pos","b.v_Neg",0.3],
    ["Neg","b.v_Pos",0.2],
    ["Neg","b.v_Neg",0.8]
],[cancer.distribution]),name="blood_vomiting")
```

```python
model=BayesianNetwork()
```

```python
model.add_states(smoking,asbes_consum,cancer,scan,blood_vomiting)
```

```python
model.add_edge(smoking,cancer)
model.add_edge(asbes_consum,cancer)
model.add_edge(cancer,scan)
model.add_edge(cancer,blood_vomiting)
```

```python
model.bake()
```

```python
model
```

```python
probability=model.probability([["Low smoking","Low cons","Pos","scan_Pos","b.v_Pos"]])
```

```python
probability
```

```python
print(model.predict([["Low smoking","Low cons","Pos","scan_Pos","b.v_Pos"]]))
```

```python
predictions=model.predict_proba({"Blood vomiting":"B.V_Pos"})
```
